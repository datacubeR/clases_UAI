```python
import pandas as pd
from sklearn.metrics import ConfusionMatrixDisplay, classification_report

df = pd.DataFrame(
    dict(
        y=[1, 1, 0, 2, 1, 0, 1, 2, 0, 1, 2],
        y_pred=[1, 0, 2, 2, 1, 0, 1, 1, 2, 1, 2],
    )
)

print(classification_report(df.y, df.y_pred, digits=3))
ConfusionMatrixDisplay.from_predictions(df.y, df.y_pred)
```

> Es importante mencionar que para el caso de más de 2 clases, el Accuracy se calcula como:

$$Accuracy = \frac{TP_{Clase\,0} + TP_{Clase\,1} + TP_{Clase\,2}}{Total} = \frac{1 + 4 + 2}{11} = \frac{7}{11} = 0.636$$

`El accuracy es un sólo valor para todas las clases. Y en el caso que mostramos ayer, el concepto de TN se pierde, por lo que es necesario reducir la matriz a un problema binario para poder poder darle sentido.`
