## Agrupar

```python
df = pd.DataFrame(
    dict(a=[1, 1, 1, 1, 2, 2, 2, 2], b=[1, 2, 3, 4, 5, 6, 7, 8])
)

df
```

```python
for i in [df.shape, df.columns, df.index, df.dtypes]:
    print(i)
```

```python
groups = df.groupby("a")
for id, g in groups:
    print(f"Este es el grupo: {id}")
    display(g)
```

```python
df.groupby("a")["b"].mean()
```

```python
titanic_df.groupby("sex")["fare"].mean()
```

```python
titanic_df.groupby(["sex", "pclass"])[["age", "fare"]].median()
```
