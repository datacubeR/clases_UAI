```python
## Varias filas y varias columnas...
titanic_df.loc[[1, 100, 890], ["who", "adult_male", "alive"]]
```

> Loc: Me sirve para poder llamar filas por su identificador (index) y también combinarlo con columnas.  
> Si quiero sólo columnas, no es necesario usar loc.  
> Si quiero filas y columnas también puedo usar loc.

```python
## Agrupamiento...

df = pd.DataFrame(
    dict(a=[1, 1, 1, 1, 2, 2, 2, 2], b=[1, 2, 3, 4, 5, 6, 7, 8])
)
df
```

```python
group = df.groupby("a")
group
```

```python
for id, g in group:
    print(f"Este es el grupo: {id}")
    display(g)
```

```python
df.groupby("a", as_index=False)["b"].mean()
```

```python
titanic_df.groupby("sex")["age"].mean()
```

```python
titanic_df.groupby("alive")["fare"].mean()
```

```python
titanic_df.groupby("pclass")[
    "survived"
].mean()  # calculo el promedio de una variable binaria...
```

```python
titanic_df.groupby(["sex", "pclass"], as_index=False)[
    ["age", "fare"]
].median()
```

```python
titanic_df.groupby(
    ["sex", "embark_town"], as_index=False
).size()  # Contar...
```

```python
titanic_df.groupby(["sex", "pclass", "embark_town", "alone"])[
    ["fare", "age"]
].mean()
```
